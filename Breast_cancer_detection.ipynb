```python
#seperating the features and target
X=df.drop(columns='target',axis=1)
y=df.iloc[:,0]
```

```python
#feature scaling
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
X=scaler.fit_transform(X)

X
```

```python
#splitting the data into training data and testing data
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
#model training
model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
#model prediction
y_pred=model.predict(X_test)
```

```python
#model Evaluation
score=accuracy_score(y_test,y_pred)
score
```

```python
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
#hyper parameter tuning
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'C': [0.01, 0.1, 1, 10, 100],             # regularization strength
    'penalty': ['l1', 'l2'],                 # regularization type
    'solver': ['liblinear']                  # works with l1 & l2
}
```

```python
model = LogisticRegression(max_iter=1000)

grid_search = GridSearchCV(estimator=model,
                           param_grid=param_grid,
                           cv=5,             # 5-fold cross-validation
                           scoring='accuracy',
                           n_jobs=-1,
                           verbose=2)
```

```python
grid_search.fit(X_train, y_train)
```

```python
print("Best Parameters:", grid_search.best_params_)
print("Best Cross-Validation Accuracy:", grid_search.best_score_)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Predict on test set
y_pred = grid_search.best_estimator_.predict(X_test)

# Accuracy
print("Test Accuracy:", accuracy_score(y_test, y_pred))

# Detailed classification report
print("\nClassification Report:\n", classification_report(y_test, y_pred))

# Confusion matrix
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
# Final Logistic Regression Model with Best Parameters
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# 1. Build the model with best parameters
logistic_model = LogisticRegression(C=10, penalty='l1', solver='liblinear', max_iter=1000)

# 2. Fit the model on training data
logistic_model.fit(X_train, y_train)

# 3. Predict on test data
y_pred = logistic_model.predict(X_test)

# 4. Evaluate the model
print(" Test Accuracy:", accuracy_score(y_test, y_pred))
print("\n Classification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```
